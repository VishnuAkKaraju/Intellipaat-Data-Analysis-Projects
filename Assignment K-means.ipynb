```python
from sklearn.cluster import KMeans
```

```python
clusters=list(range(2,8))
ssd = []   # WCSS within cluster sum of square
for num_clusters in clusters:
    model_clus = KMeans(n_clusters = num_clusters, max_iter=250,random_state= 50)
    model_clus.fit(df_df1)
    ssd.append(model_clus.inertia_) #inertial is the distance between point and centroid

plt.plot(clusters,ssd);
```

```python
cluster = KMeans(n_clusters=3, max_iter=250, random_state= 50)
cluster.fit(df_df1)
```
