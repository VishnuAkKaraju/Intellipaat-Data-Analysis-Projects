```python
from prophet import Prophet
```

```python
df = df.rename(columns={'Date': 'ds', 'Weekly_Sales': 'y','Store':'store_id'}) 
```

```python
train = df[df['ds'] < '2024-05-20'] 
```

```python
model = Prophet()
model.fit(train)
```

```python
future = model.make_future_dataframe(periods=12, freq='W')
```

```python
forecast = model.predict(future)
```

```python
forecast
```

```python
merged_df = pd.merge(left=df, right=forecast, on='ds')
print(merged_df[['ds', 'store_id', 'yhat']].head(10))
```

```python
store_forecasts =merged_df.groupby(by='store_id')
```
